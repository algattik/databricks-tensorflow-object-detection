```
dbutils.fs.put("/databricks/install-tensorflow-object-detection.sh","""#!/bin/bash
#Clone TensorFlow Object Detection API v1.12.0
mkdir -p /opt/tensorflow
cd /opt/tensorflow && git clone --branch v1.12.0 https://github.com/tensorflow/models.git

#Workaround for Python 3: bug https://github.com/tensorflow/models/issues/4780
sed -i -e 's/category_index.values(),/list(category_index.values()),/' \
  /opt/tensorflow/models/research/object_detection/model_lib.py

#Generate protoc files
cd /opt/tensorflow/models/research/
protoc object_detection/protos/*.proto --python_out=.

#Install pycocotools
git clone https://github.com/cocodataset/cocoapi.git
cd cocoapi/PythonAPI
PATH=/databricks/python/bin:$PATH make
cp -r pycocotools /opt/tensorflow/models/research/

# Install contextlib2 module
/databricks/conda/bin/conda install -y -p /databricks/python contextlib2

""", True)
```
